```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt 
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
#Using the dendrogram to find the number of clusters
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method= 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
#Training the Hierarchical model
#ward= minimum variance method
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity= 'euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print (y_hc)
```

```python
plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1], s = 70, c= 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1], s = 70, c= 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1], s = 70, c= 'black', label = 'Cluster 3')
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1], s = 70, c= 'magenta', label = 'Cluster 4')
plt.scatter(X[y_hc == 4,0],X[y_hc == 4,1], s = 70, c= 'cyan', label = 'Cluster 5')

plt.title('Hierarchical model')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Salary Score (1-100)')
plt.legend()
```
